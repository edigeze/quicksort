# Tri en comptant

```python
def counting(l):
    d=[0] * (max(l)+1) #liste de comptage des occurences etat initial
    L=[]
    for i in l: # on remplis la liste d
        d[i] = d[i]+1
    for i in range(len(d)): # on remplit la liste L à l'aide de d
        for j in range(d[i]):
            L.append(i)    
    return L
```

# Pivot sur le 1er élément et seuil de récursivité à 1

```python
def TriRapideinit(L):
    TriRapide(L,0,len(L)-1)

def TriRapide(L,start,end):
    if start>=end:
        return
    splitpoint = partition(L,start,end)
    TriRapide(L,start,splitpoint-1)# on ne garde met pas splitpoint car c'est le pivot
    TriRapide(L,splitpoint+1,end)
    
def partition(L,start,end):
    pivot = L[start]
    leftmark = start+1
    rightmark = end
    done = False

    while not done:
        while leftmark <= rightmark and L[leftmark] <= pivot:
            leftmark = leftmark + 1

        while rightmark >= leftmark and L[rightmark] >= pivot:
            rightmark = rightmark -1

        if rightmark < leftmark:
            done = True
        else:
            L[leftmark], L[rightmark] = L[rightmark],L[leftmark]
    
    # on place le pivot au milieu
    L[start],L[rightmark]=L[rightmark],L[start]
    pivot = rightmark
    return pivot
```

2.	Pivot sur le 1er élément et seuil de récursivité déterminé expérimentalement

il faut determiner le seuil pour cela on fait varier le seuil pour obtenir les meilleur resiltats


```python
def quicksortinit2(l):
    quicksort2(l, 0, len(l) - 1)
    
def quicksort2(l, first, last):
    if first<last:
        sizeArr = last - first + 1
        if(sizeArr < 15):
            insert_sort(l, first, last)
        else:
            mid = partition2(l, first, last)
            quicksort2(l, first, mid-1)
            quicksort2(l, mid + 1, last)

def partition2(l, first, last):
    piv = l[last]
    i = first-1
    for j in range(first,last):
        if l[j] < piv:
            i += 1
            l[i],l[j] = l[j],l[i]

    l[i+1],l[last] = l[last],l[i+1]
    return i+1

def insert_sort(l, first, last):
    for x in range(first, last+1):
        key = l[x]
        y = x-1
        while y > -1 and l[y]> key:
            l[y+1] = l[y]
            y = y-1
        l[y+1] = key
```

3.	Pivot sur un élément aléatoire et seuil de récursivité déterminé expérimentalement

```python
def quicksortinit3(l):
    quicksort3(l, 0, len(l) - 1)
    
def quicksort3(l, first, last):
    if first<last:
        sizeArr = last - first + 1
        if(sizeArr < 15):
            insert_sort(l, first, last)
        else:
            mid = partition3(l, first, last)
            quicksort3(l, first, mid-1)
            quicksort3(l, mid + 1, last)

def partition3(l, first, last):
    att = random.randint(first,last)
    l[att],l[last] = l[last],l[att]# on prend un pivot au hasard et on le met au debut
    piv = l[last]
    i = first-1
    for j in range(first,last):
        if l[j] < piv:
            i += 1
            l[i],l[j] = l[j],l[i]

    l[i+1],l[last] = l[last],l[i+1]
    return i+1

def insert_sort(l, first, last):
    for x in range(first, last+1):
        key = l[x]
        y = x-1
        while y > -1 and l[y]> key:
            l[y+1] = l[y]
            y = y-1
        l[y+1] = key
```
